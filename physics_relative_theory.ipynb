## Special Relativity

Special relativity, developed by Albert Einstein in 1905, describes how space and time behave when objects move at speeds close to the speed of light. It is based on two postulates:
1. The laws of physics are the same in all inertial frames of reference.
2. The speed of light in a vacuum is constant for all observers, regardless of their motion.

### 1. Time Dilation (The Twin Paradox)
Time dilation occurs when an observer moving relative to a stationary observer experiences time more slowly.

The time dilation formula is:

$$ t' = \frac{t}{\gamma} $$

where:
- \( t' \) is the time experienced by the moving observer (spaceship twin).
- \( t \) is the time experienced by the stationary observer (Earth twin).
- \( Î³ \) is the Lorentz factor, given by:

$$ \gamma = \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}} $$

where:
- \( v \) is the velocity of the moving object.
- \( c \) is the speed of light.

If a twin travels at 90% the speed of light for 10 years (Earth time), their aging time would be:

$$ t' = \frac{10}{\gamma} = \frac{10}{\frac{1}{\sqrt{1 - 0.9^2}}} \approx 4.36 \text{ years} $$

So the traveling twin ages only 4.36 years, while the Earth twin ages 10 years.

### 2. Length Contraction
An object moving at relativistic speeds appears shorter in the direction of motion.

The formula for length contraction is:

$$ L = L_0 \sqrt{1 - \frac{v^2}{c^2}} $$

where:
- \( L_0 \) is the proper length (length in the object's own frame).
- \( L \) is the contracted length observed from Earth.
- \( v \) is the object's velocity.

For example, if a spaceship is 100 meters long at rest and moves at 90% the speed of light:

$$ L = 100 \times \sqrt{1 - 0.9^2} \approx 43.6 \text{ meters} $$

From Earth, the spaceship appears much shorter.


```python
### Special and General Relativity Examples

## Special Relativity

### 1. Time Dilation (The Twin Paradox)
# Time dilation occurs when an observer moving relative to a stationary observer experiences time more slowly.
# Imagine two twins: one stays on Earth while the other travels in a spaceship close to the speed of light.
# The traveling twin will age more slowly compared to the one on Earth.

import numpy as np
import matplotlib.pyplot as plt


def time_dilation(proper_time, velocity, c=3e8):
    gamma = 1 / np.sqrt(1 - (velocity**2 / c**2))  # Lorentz factor
    dilated_time = proper_time * gamma
    return dilated_time


# Example: Compare time dilation for different velocities
velocities = np.linspace(0, 0.99 * 3e8, 100)  # From 0 to 99% speed of light
proper_time = 5  # 5 years in stationary frame
dilated_times = time_dilation(proper_time, velocities)

plt.plot(velocities / 3e8, dilated_times, label="Time Experienced in Spaceship")
plt.axhline(y=proper_time, color="r", linestyle="--", label="Time Experienced on Earth")
plt.xlabel("Velocity (fraction of c)")
plt.ylabel("Time Experienced (years)")
plt.title("Time Dilation: Twin Paradox")
plt.legend()
plt.grid()
plt.show()

### 2. Length Contraction (Spaceship Traveling at High Speed)
# If a spaceship moves at high speed, an observer on Earth sees it as shorter than its actual length.


def length_contraction(proper_length, velocity, c=3e8):
    gamma = 1 / np.sqrt(1 - (velocity**2 / c**2))
    contracted_length = proper_length / gamma
    return contracted_length


# Example: Compare length contraction at different velocities
proper_length = 100  # 100 meters spaceship
contracted_lengths = length_contraction(proper_length, velocities)

plt.plot(velocities / 3e8, contracted_lengths, label="Contracted Length of Spaceship")
plt.axhline(
    y=proper_length, color="r", linestyle="--", label="Actual Length of Spaceship"
)
plt.xlabel("Velocity (fraction of c)")
plt.ylabel("Length (meters)")
plt.title("Length Contraction in Special Relativity")
plt.legend()
plt.grid()
plt.show()
```

## General Relativity

General relativity, proposed by Albert Einstein in 1915, describes how gravity affects the fabric of space and time. It states that massive objects warp spacetime, affecting the motion of objects and even the flow of time.

### 1. Gravitational Time Dilation
Time runs slower in a strong gravitational field. This means that an astronaut near a massive planet experiences time more slowly than someone far away.

The gravitational time dilation formula is:

$$ t' = t \sqrt{1 - \frac{2GM}{rc^2}} $$

where:
- \( t' \) is the time experienced near the massive object.
- \( t \) is the time far from the massive object.
- \( G \) is the gravitational constant.
- \( M \) is the mass of the object (e.g., a planet).
- \( r \) is the distance from the center of the mass.
- \( c \) is the speed of light.

For example, near Earth (mass $ 5.97 \times 10^{24} $ kg) at a distance \( r = 10,000 \) km:

$$ t' = t \sqrt{1 - \frac{2(6.674 \times 10^{-11})(5.97 \times 10^{24})}{(10^7)(3 \times 10^8)^2}} $$

This effect is measured in satellites, requiring GPS clocks to be adjusted for time dilation.

### 2. Gravitational Lensing (Light Bending)
Gravity bends light, causing distant objects to appear distorted when viewed near a massive object.

The Einstein radius (angle of deflection) is given by:

$$ \theta_E = \sqrt{\frac{4GM}{c^2 d}} $$

where:
- \( ð¤_E \) is the Einstein radius (bending angle).
- \( d \) is the distance from the lensing object.

For example, if light from a distant star passes near a massive planet, the light bends, making the star appear shifted.

This effect is used in astronomy to detect exoplanets and distant galaxies.



```python
## General Relativity

### 1. Gravitational Time Dilation (Time Runs Differently in Space)
# Imagine an astronaut traveling in a spaceship near a massive planet, while another stays far from the planet.
# The astronaut near the planet experiences slower time compared to the one farther away.


def gravitational_time_dilation(mass, radius, c=3e8, G=6.67430e-11):
    Rs = 2 * G * mass / c**2  # Schwarzschild radius
    return 1 / np.sqrt(1 - Rs / radius)


# Example: Compare time dilation near a massive planet vs far away
mass = 6e24  # Approximate mass of Earth
radii = np.linspace(1e6, 1e8, 100)  # Distance from center of planet
time_dilations = gravitational_time_dilation(mass, radii)

plt.plot(radii / 1e3, time_dilations, label="Time Dilation Factor")
plt.yscale("log")  # Logarithmic scale for better visualization
plt.xlabel("Distance from Planet (km)")
plt.ylabel("Time Dilation Factor")
plt.title("Gravitational Time Dilation: Space vs Planet Surface")
plt.legend()
plt.grid()
plt.show()

### 2. Gravitational Lensing (Light Bending Near a Massive Object)
# If a spaceship passes near a massive planet, the light from distant stars appears distorted.


def einstein_radius(mass, distance, c=3e8, G=6.67430e-11):
    return np.sqrt(4 * G * mass / (c**2 * distance))


# Example: Compute Einstein radius for different planet masses
distances = np.linspace(1e16, 1e18, 100)  # Distance from lensing object
masses = [6e24, 1e30]  # Earth and Sun mass
radii = [einstein_radius(m, distances) for m in masses]

plt.plot(distances / 1e16, radii[0], label="Earth Mass")
plt.plot(distances / 1e16, radii[1], label="Sun Mass")
plt.xlabel("Distance from Lens (10^16 m)")
plt.ylabel("Einstein Radius (radians)")
plt.yscale("log")  # Logarithmic scale for better visualization
plt.title("Gravitational Lensing: Light Bending Effect")
plt.legend()
plt.grid()
plt.show()
```

## Maxwell's Equations and the Birth of Relativity

Maxwell's equations unified electricity, magnetism, and light into a single theoretical framework. They predict that light is an electromagnetic wave propagating at a constant speed $c$, independent of the observer's frame of reference. This key insight directly inspired Einstein's Special Theory of Relativity.

### Derivation of the Electromagnetic Wave Equation

In a vacuum with no charges ($\rho = 0$) or currents ($\mathbf{J} = 0$), Maxwell's equations simplify to:

1. **Gauss's Law**:  
   $$ \nabla \cdot \mathbf{E} = 0 $$

2. **Gauss's Law for Magnetism**:  
   $$ \nabla \cdot \mathbf{B} = 0 $$

3. **Faraday's Law**:  
   $$ \nabla \times \mathbf{E} = -\frac{\partial \mathbf{B}}{\partial t} $$

4. **AmpÃ¨re's Law (with Maxwell's correction)**:  
   $$ \nabla \times \mathbf{B} = \mu_0 \epsilon_0 \frac{\partial \mathbf{E}}{\partial t} $$

**Wave Equation Derivation**:  
1. Take the curl of Faraday's Law:  
   $$ \nabla \times (\nabla \times \mathbf{E}) = -\frac{\partial}{\partial t} (\nabla \times \mathbf{B}) $$

2. Apply the vector identity $\nabla \times (\nabla \times \mathbf{E}) = \nabla (\nabla \cdot \mathbf{E}) - \nabla^2 \mathbf{E}$. Using $\nabla \cdot \mathbf{E} = 0$:  
   $$ -\nabla^2 \mathbf{E} = -\frac{\partial}{\partial t} \left( \mu_0 \epsilon_0 \frac{\partial \mathbf{E}}{\partial t} \right) $$

3. Simplify to obtain the wave equation for $\mathbf{E}$:  
   $$ \nabla^2 \mathbf{E} = \mu_0 \epsilon_0 \frac{\partial^2 \mathbf{E}}{\partial t^2} $$

4. Similarly, taking the curl of AmpÃ¨re's Law yields an identical wave equation for $\mathbf{B}$:  
   $$ \nabla^2 \mathbf{B} = \mu_0 \epsilon_0 \frac{\partial^2 \mathbf{B}}{\partial t^2} $$

The wave speed is:  
$$ c = \frac{1}{\sqrt{\mu_0 \epsilon_0}} $$

### Implications for Relativity
1. **Constant Speed of Light**:  
   The speed $c$ depends only on vacuum properties ($\mu_0$, $\epsilon_0$) and is independent of the observerâs motion. This contradicted Galilean relativity, where velocities add linearly.

2. **Einsteinâs Resolution**:  
   Einstein resolved this paradox by proposing that space and time are relative, not absolute. This led to the **Special Theory of Relativity** (1905), where $c$ is invariant in all inertial frames.

This derivation showed that light is an electromagnetic wave with a fixed speed, a cornerstone of modern physics and the foundation of relativity.
